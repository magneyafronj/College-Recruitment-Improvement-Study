### Handling null values in the Impact columns
#### Used AI to figure out how to do this

```python
impact_cols = [col for col in df.columns if 'impact' in col or 'attend_factor' in col]
df[impact_cols] = df[impact_cols].fillna('No Impact')

interest_topic_cols = [col for col in df.columns if 'course_interest' in col]
df[interest_topic_cols] = df[interest_topic_cols].fillna('No')
print(df.isnull().sum().sum())
```

### Grouping major and Race
#### Used AI to figure out how to do this

```python
def categorize_major(m):
    m = str(m).lower().strip()
    if any(x in m for x in ['nursing', 'radiography']): return 'health'
    if any(x in m for x in ['psychology', 'child', 'education', 'family']): return 'social_science_education'
    if any(x in m for x in ['broadcasting', 'journalism', 'communication', 'theatre', 'music', 'culinary', 'arts']): return 'arts_comm'
    if any(x in m for x in ['engineering', 'biology', 'science', 'analyst']): return 'stem'
    if any(x in m for x in ['liberal arts', 'history', 'justice']): return 'liberal_arts'
    return 'other_non_degree'

df['degree_program'] = df['degree_program'].apply(categorize_major)

# Cleaning and condensing 'race_ethnicity'
def simplify_race(r):
    r = str(r)
    if ';' in r: return 'multi_racial'
    if 'reply' in r.lower(): return 'unknown'
    return r.lower().replace(' ', '_').replace('/', '_')

df['race_ethnicity'] = df['race_ethnicity'].apply(simplify_race)
```
