## Dataframes Introduction

The pandas Dataframe is a very useful and efficient data structure for data analysis. It is commonly imported as pd.

```python
import pandas as pd
```

A common data file format is Comma Separated Value (CSV) files. Use the read_csv() method to bring it into a Python script.
* Note the dataset weather.csv was pulled from the NOAA database for the weather stations in/near Morris County New Jersey.  The file should be stored in the same folder as this notebook.  Or alternatively, include the full pathname of the file in the cell below.

```python
datafile = "weather.csv"
# Read in the file and set column 2 (the DATE) as the index and looks for fields that looks like dates
df = pd.read_csv(datafile, index_col=2,parse_dates=True)  

# An alternative read which will use index numbers beginning at 0 and sets a specific column to be a date
df2 = pd.read_csv(datafile, parse_dates=['DATE'])
```

```python
df #notice that the first column is the index and that the index is a date, not an integer 
```

```python
df2  # notice that the index is an integer
```

## Selecting Subsets of a Dataframe

Narrow down your analysis by selecting portions of your dataframe

### Select specific columns

#### Select a single column

```python
df.NAME
```

#### Select a single column (alternate form - use for column names that aren't a single word)

```python
df['STATION']
```

#### Select a subset of the dataframe (multiple columns)

```python
rainandsnow = df[['STATION','PRCP','SNOW']]
rainandsnow
```

```python
df.PRCP.mean()
```

```python
df.STATION.unique()
```

```python
df.STATION.value_counts()
```

Select a range of columns by names using loc[]

```python
df.loc[:,'TAVG':'TMIN']
```

#### Filter dataframe by a condition

```python
rainandsnow_station = rainandsnow[rainandsnow.STATION == 'US1NJOC0058']
rainandsnow_station
```

#### reset_index() changes index back to an integer and makes the former index a column

```python
rainandsnow_station.reset_index(inplace=True)
rainandsnow_station
```

#### Filter data multiple conditions 

```python
raindays = df[(df['PRCP']>0) & (df['STATION'] == 'US1NJOC0058')]
raindays
```

```python
raindays.reset_index(inplace=True)
raindays
```
